```python
import fitz

pdf = fitz.open("../Stock Trader's Almanac 2026_L.pdf")

print("Pages:", len(pdf))
```

```python
for page_num in range(len(pdf)):

    text = pdf[page_num].get_text()

    if "JUNE ALMANAC" in text:

        print(
            "Found on page:",
            page_num + 1
        )
```

```python
for page_num in range(79, 85):

    print("\n")
    print("=" * 80)
    print("PAGE", page_num + 1)
    print("=" * 80)

    text = pdf[page_num].get_text()

    print(text[:5000])
```

```python
for page_num in range(79, 85):

    page = pdf[page_num]

    pix = page.get_pixmap(
        matrix=fitz.Matrix(3,3)
    )

    pix.save(
        f"june_page_{page_num+1}.png"
    )

print("Saved")
```
