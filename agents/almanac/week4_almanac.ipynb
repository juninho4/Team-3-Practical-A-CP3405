```python
import pandas as pd


def analyze_month(
    month,
    avg_return,
    midterm_return
):

    if midterm_return < 0:
        verdict = "Bearish"
    elif midterm_return > 1:
        verdict = "Bullish"
    else:
        verdict = "Neutral"

    return pd.DataFrame({
        "Month":[month],
        "Average Return":[avg_return],
        "Midterm Return":[midterm_return],
        "Verdict":[verdict]
    })
```

```python
may = analyze_month(
    "May",
    0.3,
    -0.7
)

may
```

```python
june = analyze_month(
    "June",
    0.2,
    -2.1
)

june
```
